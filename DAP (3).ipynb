```python
#mathematical operations
a=np.array([1,2,3,4])
b=np.array([4,5,6,7])
s2=np.array([1,2,3,4,5,6])
c=np.add(a,b)
print(c)
#broad casting
d=np.add(a,3)
print(d)
#array manipulation
s=np.vstack(a+b)
print(s)
e=np.stack(a+b)
print(e)
s1=np.split(s2,2)
print(s1)
```

```python
#transpose
y=np.array([[1,2,3,4],[5,6,7,8]])
b=y.T
print(b)
```

```python
z=np.array([[1,2],[3,4]])
x=np.array([[5,6],[7,8]])
s=np.dot(z,x)
print(s)
e=np.linalg.eig(s)
print(e)
```

```python
import matplotlib.pyplot as plt
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='CompanyA',marker='>',ms='15',mec='b')
plt.legend(loc='best')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='CompanyB',marker='H')
plt.legend(loc='best')
```
