**Step:-3-Importing the required libraries.**

```python
import pandas as pd
from selenium import webdriver
from time import sleep
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

**Step:-4-Defining options and setting browser capabilities
        creating webdriver instance with options
        Get the url
        Connecting with the IMDB website using .get function**

```python
#Define options and set browser capabilities
options = webdriver.ChromeOptions()
options.add_argument('--some--option')
#create webdriver instance with options
driver=webdriver.Chrome(options=options)
driver.get("https://m.imdb.com/search/title/?title_type=feature&genres=horror")
```

**Step:-5-Loading the following data from the website.
            1.Titles
            2.Ratings
            3.Views
            4.Summary
        Save the data in the program using list comprehension.**

```python
a=driver.find_elements(By.XPATH,".//a[@class='ipc-title-link-wrapper']")
Titles=[i.text for i in a]
print(Titles)
print("Total No. of titles=",len(Titles))
```

```python
b=driver.find_elements(By.XPATH,".//span[@class='ipc-rating-star--voteCount']")
rating=[i.text for i in b]
print(rating)
print(len(rating))
```

```python
c=driver.find_elements(By.XPATH,".//div[@class='sc-f24f1c5c-0 cPpOqU']")
votes=[i.text for i in c]
print(votes)
```

```python
d=driver.find_elements(By.XPATH,".//div[@class='ipc-html-content-inner-div']")
sum=[i.text for i in d]
print(sum)
```

**Step:-6-Creating a dataframe and add the contents to dataframe.**

```python
df=pd.DataFrame(list(zip(Titles,rating,votes,sum)),columns=['Titles','Rating','Votes','Summary'])
print(df)
```
